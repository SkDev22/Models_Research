```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
df = pd.read_csv("Dataset_Cleaned.csv")
df.columns = df.columns.str.strip()
```

```python
plt.figure(figsize=(10, 5))
df['Location'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Number of Boarding Houses by Location')
plt.xlabel('Location')
plt.ylabel('Number of Listings')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
df['Price (LKR)'].plot(kind='hist', bins=20, color='orange', edgecolor='black')
plt.title('Price Distribution of Boarding Houses')
plt.xlabel('Price (LKR)')
plt.ylabel('Number of Listings')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(6, 6))
df['Room_Type'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=sns.color_palette('Set2'))
plt.title('Room Type Distribution')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
df['Availability'].value_counts().plot(kind='bar', color='lightgreen')
plt.title('Availability Status of Boarding Houses')
plt.xlabel('Availability')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
from collections import Counter

# Convert amenities to list
amenities_list = df['Amenities'].dropna().str.lower().str.split(', ')
all_amenities = [amenity for sublist in amenities_list for amenity in sublist]
top_amenities = Counter(all_amenities).most_common(10)

# Create DataFrame for chart
amenities_df = pd.DataFrame(top_amenities, columns=['Amenity', 'Count'])

plt.figure(figsize=(8, 5))
sns.barplot(data=amenities_df, x='Amenity', y='Count', palette='Blues_d')
plt.title('Top 10 Most Common Amenities')
plt.xticks(rotation=45)
plt.show()
```
